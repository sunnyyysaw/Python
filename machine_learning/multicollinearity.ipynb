```python
import pandas as pd
```

```python
import statsmodels.api as sm
df = pd.read_csv("D:/dataset/kn_ml/Advertising.csv", index_col = 0)
df
```

```python
X = df[['TV', 'radio', 'newspaper']]
y = df[['sales']]
```

```python
X = sm.add_constant(X)  # Add constant term (intercept)
ols_model = sm.OLS(y, X).fit()
```

```python
ols_model.summary()
```

```python
import matplotlib.pyplot as plt
X.iloc[:, 1:].corr()
```
