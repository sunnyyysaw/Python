```python
dff = pd.merge(df, df1, on = 'Country Code', how = 'left')
dff
```

```python
dff.dtypes
```

```python
coun_names = dff.Country.value_counts().index
coun_val = dff.Country.value_counts().values
```

```python
# top 3 countries that uses zomato
plt.pie(coun_val[:3], labels = coun_names[:3], autopct = '%1.2f%%')
plt.show()
```

zomato maximum transactions/records are from India and then followed by USA and then UK

```python
dff.columns
```

```python
dff.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()
```

1. when ratings are between 4.5 and 4.9 ---> it indicates customer has excellent satisfaction
2. when the rating are between 4.0 and 4.4 ---> very good satisfaction
3. and when the rating are between 3.5 and 3.9 ---> good satisfaction
4. next when the rating are between 2.5 and 3.4 ---> average satisfaction and so on...

```python
ratings = dff.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns = {0 : 'Rating Count'})
```

```python
ratings.head()
```

```python
plt.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x = 'Aggregate rating', y = 'Rating Count', data = ratings)
plt.show()
```

```python
sns.barplot(x = 'Aggregate rating', y = 'Rating Count', data = ratings, hue = 'Rating color')
plt.show()
```

```python
sns.barplot(x = 'Aggregate rating', y = 'Rating Count', hue = 'Rating color', data = ratings, palette = ['White', 'Red', 'Orange', 'Yellow', 'Green', 'Green'])
plt.show()
```

1. NOT RATED count is very high
2. Maximum ratings count lies between range of 2.5 to 4.5

```python
sns.countplot(x = 'Rating color', data = ratings, palette = ['White', 'Red', 'Orange', 'Yellow', 'Green', 'Green'])
plt.show()
```

```python
# find the countries name that has given 0 rating
dff[dff['Rating color'] == 'White'].groupby('Country').size().reset_index()
```

Maximum number of 0 ratings are from Indian customers
